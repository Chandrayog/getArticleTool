```python
import pandas as pd
import numpy as np

### Plots
import seaborn as sns
# import matplotlib.pyplot as plt
# import plotly.offline as py



#ignore warning messages
import warnings
warnings.filterwarnings('ignore')


### setting output display options
pd.set_option('display.width', 400)
pd.set_option('display.max_columns', 10)

### Load the dataset
df = pd.read_excel("/Users/chandra/Desktop/WA-Fn-UseC--HR-Employee-Attrition.xlsx")

# Summary of the records
print("This dataset has {} samples with {} features each.".format(df.shape[0], df.shape[1]))


# Display the first 10 record
print("\nDisplay the first 10 record")
print(df.head(10))

# Describe the data
print("Statistical description of dataset\n--------------------------------------")
print(df.describe())


# count of employees in different departments
sns.set_theme(style="whitegrid")
sns.histplot(data=df, x="Department")


# Load the dataset and draw Scatterplot for AGE wise Travel Details of Emp based on their YearsinCompany and Department
cmap = sns.cubehelix_palette(rot=-.2, as_cmap=True)
g = sns.relplot(
    data=df,
    x="BusinessTravel", y="Age",
    hue="YearsAtCompany", size="Department",
    palette=cmap, sizes=(10, 200),
)

g.ax.xaxis.grid(True, "minor", linewidth=.25)
g.ax.yaxis.grid(True, "minor", linewidth=.25)
g.despine(left=True, bottom=True)


# WorklifeBalance based on the YearsSinceLastPromotion
sns.set_theme(style="darkgrid")
sns.displot(df, x="WorkLifeBalance", hue="YearsSinceLastPromotion", element="step")

```
